```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# Gantilah 'nama_file.csv' dengan nama file CSV yang sesuai dengan data Anda
data = pd.read_csv('HIV_AIDS.csv')
kolom_yang_diinginkan = data[['UHF', 'Age']]
# Contoh: Menggunakan one-hot encoding dengan Pandas
data_encoded = pd.get_dummies(kolom_yang_diinginkan, drop_first=True)
korelasi = data_encoded.corr()
plt.figure(figsize=(8, 6))  # Menentukan ukuran heatmap
sns.heatmap(korelasi, annot=True, cmap='coolwarm', fmt=".2f")

# Menambahkan xlabel, ylabel, dan title
plt.xlabel('Kategori A')
plt.ylabel('Kategori B')
plt.title('Heatmap Korelasi antara Kategori A dan Kategori B')

# Menampilkan heatmap
plt.show()
```
