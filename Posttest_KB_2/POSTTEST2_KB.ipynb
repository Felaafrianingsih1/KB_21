```python

import pandas as pd

# Baca dataset dari file CSV
df = pd.read_csv('HIV_AIDS.csv')
df.head()
 
```

```python
#soal no 1
# Menghitung rata-rata
rata_rata = df['HIV diagnoses'].mean()

# Menghitung standar deviasi
standar_deviasi = df['HIV diagnoses'].std()

# Menghitung nilai minimum
nilai_minimum = df['HIV diagnoses'].min()

# Menghitung nilai maksimum
nilai_maksimum = df['HIV diagnoses'].max()

print("Rata-rata:")
print(rata_rata)

print("\nStandar Deviasi:")
print(standar_deviasi)

print("\nNilai Minimum:")
print(nilai_minimum)

print("\nNilai Maksimum:")
print(nilai_maksimum)

```

```python
#soal no 2
# Filter kolom-kolom yang hanya bertipe angka
kolom_angka = df.select_dtypes(include=['int', 'float'])

# Menampilkan 10 record pertama dari dataset
kolom_angka= df[['Year', 'HIV diagnoses','HIV diagnosis rate','Concurrent diagnoses','% linked to care within 3 months','AIDS diagnoses','AIDS diagnosis rate','PLWDHI prevalence','% viral suppression','Deaths','Death rate','HIV-related death rate','Non-HIV-related death rate']]
df.head(10)
```

```python
#soal no 3
# Informasi jumlah rekaman
jumlah_rekaman = len(df)

# Informasi jumlah atribut
jumlah_atribut = len(df.columns)

# Nama atribut
nama_atribut = df.columns.tolist()

# Informasi jumlah atribut dengan tipe angka
jumlah_atribut_angka = len(df.select_dtypes(include=['int', 'float']).columns)

# Menyusun informasi dalam dictionary
informasi_dataset = {
    'Jumlah Rekaman': jumlah_rekaman,
    'Jumlah Atribut': jumlah_atribut,
    'Nama Atribut': nama_atribut,
    'Jumlah Atribut dengan Tipe Angka': jumlah_atribut_angka
}

# Menampilkan isi dari dictionary
for kunci, nilai in informasi_dataset.items():
    print(f"{kunci}: {nilai}")
```
